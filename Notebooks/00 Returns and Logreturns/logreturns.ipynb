## üßÆ 1. **Definiciones Formales**

Sea $P_t$ el precio de un activo financiero en el tiempo $t$. Entonces:

* **Retorno simple (porcentaje de cambio):**

$$

* **Log-retorno (retorno continuo o compuesto continuamente):**

Ambos miden el rendimiento de un activo, pero tienen propiedades anal√≠ticas distintas.

## üß† 2. **Propiedades Matem√°ticas**

### a. **Aditividad temporal** (clave para modelos estoc√°sticos)

  $$

### ‚öôÔ∏è **Aditividad temporal: una disecci√≥n rigurosa**

La **aditividad temporal** se refiere a la propiedad estructural de ciertas transformaciones o m√©tricas financieras que **preservan la suma a lo largo del tiempo**, es decir, que el valor acumulado de una magnitud medida en sucesivos intervalos de tiempo es **igual a la suma de sus valores individuales**.

Formalmente, una funci√≥n $f(P_t, P_{t-1})$ es **aditiva en el tiempo** si cumple:

$$
\sum_{t=1}^{n} f(P_t, P_{t-1}) = f(P_n, P_0)
$$

---

### üßÆ **Log-retornos como cantidad aditiva en el tiempo**

El **log-retorno**, definido como:

$$
r_t = \log\left(\frac{P_t}{P_{t-1}}\right),
$$

cumple la propiedad:

$$
\sum_{t=1}^n r_t = \sum_{t=1}^n \log\left(\frac{P_t}{P_{t-1}}\right) = \log\left(\frac{P_1}{P_0}\right) + \log\left(\frac{P_2}{P_1}\right) + \cdots + \log\left(\frac{P_n}{P_{n-1}}\right)
$$

Por propiedades logar√≠tmicas:

$$
= \log\left( \frac{P_n}{P_0} \right)
$$

Es decir:

$$
\boxed{\sum_{t=1}^n r_t = \log\left(\frac{P_n}{P_0}\right)}
$$

---

### üìâ **Retornos simples no son aditivos**

Si usamos el retorno simple:

$$
R_t = \frac{P_t - P_{t-1}}{P_{t-1}} = \frac{P_t}{P_{t-1}} - 1,
$$

entonces:

$$
\sum_{t=1}^n R_t \neq \frac{P_n - P_0}{P_0} = R_{0 \to n}
$$

Ejemplo concreto:

#### Sup√≥n:

$$
P_0 = 100,\quad P_1 = 110,\quad P_2 = 121
$$

* Retorno simple de $t = 0$ a $t = 1$:

$$

$$

* Retorno simple de $t = 1$ a $t = 2$:

$$
R_2 = \frac{121 - 110}{110} = 0.10
$$

* Retorno **acumulado** (total):

$$
\frac{121 - 100}{100} = 0.21 \neq R_1 + R_2 = 0.20

Pero en log-retornos:

* $r_1 = \log(110 / 100) = \log(1.1) \approx 0.09531$

\boxed{\text{Suma de log-retornos } = \log\left( \frac{P_2}{P_0} \right)}
