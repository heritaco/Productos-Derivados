$S_t=S_0\,e^{\text{algo}}$ con $\text{algo}=$ **retorno log acumulado**.

**Discreto**

* Define $r_i=\ln\!\left(\frac{S_i}{S_{i-1}}\right)$.
* Entonces

$$
\ln\!\left(\frac{S_t}{S_0}\right)=\sum_{i=1}^{t} r_i
\quad\Rightarrow\quad
S_t=S_0\,\exp\!\Big(\sum_{i=1}^{t} r_i\Big).
$$

Código: `S = S0*np.exp(np.cumsum(log_returns))`.

**Si partías de retornos aritméticos $R_i=\tfrac{S_i}{S_{i-1}}-1$**

$$
S_t=S_0\prod_{i=1}^{t}(1+R_i)
= S_0\,\exp\!\Big(\sum_{i=1}^{t}\ln(1+R_i)\Big),
$$

y $\,r_i=\ln(1+R_i)$.

**GBM continuo**

$$
\frac{dS_t}{S_t}=\mu\,dt+\sigma\,dW_t
\;\Rightarrow\;
\ln S_t=\ln S_0+(\mu-\tfrac12\sigma^2)t+\sigma W_t,
$$

$$
S_t=S_0\,\exp\!\big((\mu-\tfrac12\sigma^2)t+\sigma W_t\big).
$$

Discretización exacta por pasos $\Delta t$:

$$
S_{k+1}=S_k\exp\!\big((\mu-\tfrac12\sigma^2)\Delta t+\sigma\sqrt{\Delta t}\,\varepsilon_k\big),\quad \varepsilon_k\sim\mathcal N(0,1).
$$


$S_t=S_0\,\exp\!\big((\mu-\tfrac12\sigma^2)t+\sigma W_t\big)$.

**Derivación paso a paso**

1. **Ecuación**. GBM:

$$
\frac{dS_t}{S_t}=\mu\,dt+\sigma\,dW_t
\quad\Leftrightarrow\quad
dS_t=\mu S_t\,dt+\sigma S_t\,dW_t.
$$

2. **Función**. Tome $f(s)=\ln s$. Derivadas: $f'(s)=1/s$, $f''(s)=-1/s^2$.

3. **Itô**. Para $X_t=S_t$:

$$
df(X_t)=f'(S_t)\,dS_t+\tfrac12 f''(S_t)\,(dS_t)^2.
$$

4. **Cálculo de $(dS_t)^2$**. Con $dS_t=\mu S_t\,dt+\sigma S_t\,dW_t$:

$$
(dS_t)^2=(\sigma S_t\,dW_t)^2=\sigma^2 S_t^2\, (dW_t)^2=\sigma^2 S_t^2\,dt,
$$

porque $(dt)^2=dt\,dW_t=0$ y $(dW_t)^2=dt$.

5. **Sustituya** en Itô:

$$
d\ln S_t=\frac{1}{S_t}(\mu S_t\,dt+\sigma S_t\,dW_t)+\tfrac12\!\left(-\frac{1}{S_t^2}\right)\!(\sigma^2 S_t^2\,dt).
$$

Simplifique:

$$
d\ln S_t=(\mu-\tfrac12\sigma^2)\,dt+\sigma\,dW_t.
$$

6. **Integre** de $0$ a $t$ y use $W_0=0$:

$$
\ln S_t-\ln S_0=(\mu-\tfrac12\sigma^2)t+\sigma W_t.
$$

7. **Exponentie**:

$$
S_t=S_0\,\exp\!\big((\mu-\tfrac12\sigma^2)t+\sigma W_t\big).
$$

**Consecuencias útiles**

* **Log-retorno**:

$$
\underbrace{\ln(S_t/S_0)}_{\text{log-retorno acumulado}}
=(\mu-\tfrac12\sigma^2)t+\sigma W_t\sim\mathcal N\!\big((\mu-\tfrac12\sigma^2)t,\;\sigma^2 t\big).
$$

* **Media y varianza**:

$$
\mathbb E[S_t]=S_0 e^{\mu t},\qquad
\operatorname{Var}(S_t)=S_0^2 e^{2\mu t}\big(e^{\sigma^2 t}-1\big).
$$

* **Simulación exacta por pasos $\Delta t$**:

$$
S_{k+1}=S_k\exp\!\big((\mu-\tfrac12\sigma^2)\Delta t+\sigma\sqrt{\Delta t}\,\varepsilon_k\big),
\quad \varepsilon_k\sim\mathcal N(0,1)\ \text{i.i.d.}
$$
